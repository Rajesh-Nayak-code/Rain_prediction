```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier,StackingClassifier,AdaBoostClassifier,GradientBoostingClassifier
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("rain.csv")
df.isna().sum()
df["Rain"]=df["Rain"].map({"rain":1,"no rain":0})
print(df.head(5))
```

```python
X=df.drop(columns="Rain")
y=df["Rain"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
# Model using logistic regression
model1=LogisticRegression(max_iter=200)
model1.fit(X_train,y_train)
y_pred=model1.predict(X_test)
print(f"accuracy:{accuracy_score(y_test,y_pred)}")
```

```python
# Model using KNN
model2=KNeighborsClassifier(n_neighbors=3)
model2.fit(X_train,y_train)
y_pred=model2.predict(X_test)
print(f"accuracy:{accuracy_score(y_test,y_pred)}")
```

```python
# Model using SVM
model3=SVC()
model3.fit(X_train,y_train)
y_pred=model3.predict(X_test)
print(f"accuracy:{accuracy_score(y_test,y_pred)}")
```

```python
# Model using Decision tree
model4=DecisionTreeClassifier()
model4.fit(X_train,y_train)
y_pred=model4.predict(X_test)
print(f"accuracy:{accuracy_score(y_test,y_pred)}")
```

```python
# Model using Naive bayes
model5=GaussianNB()
model5.fit(X_train,y_train)
y_pred=model5.predict(X_test)
print(f"accuracy:{accuracy_score(y_test,y_pred)}")
```

```python
score=cross_val_score(model1,X,y,cv=5)
print(f"fold_score:{score}")
print(f"avg_cross_val_score:{score.mean():.4f}")
print(f"std_score:{score.std():.4f}")
```

```python
parameters={"kernel":['linear', 'poly', 'rbf', 'sigmoid']}
grid_search=GridSearchCV(model3,parameters,cv=5,scoring="accuracy")
grid_search.fit(X, y)
print("Best Parameters:", grid_search.best_params_)
print("Best Score:", grid_search.best_score_)
```

```python
parameters={"kernel":['linear', 'poly', 'rbf', 'sigmoid']}
random_search=RandomizedSearchCV(model3,parameters,cv=5,scoring="accuracy")
random_search.fit(X, y)
print("Best Parameters:", random_search.best_params_)
print("Best Score:", random_search.best_score_)
```

```python
# Ensemble Learning

# stacking
estimators=[("dt",DecisionTreeClassifier()),("svc",SVC())]
stacking=StackingClassifier(estimators=estimators,final_estimator=LogisticRegression())
stacking.fit(X_train,y_train)
print(stacking.score(X_test,y_test))
```

```python
# Boosting ensemble learning

gradient_boost=GradientBoostingClassifier()
gradient_boost.fit(X_train,y_train)
print(gradient_boost.score(X_test,y_test))

ada_boost=AdaBoostClassifier()
ada_boost.fit(X_train,y_train)
print(ada_boost.score(X_test,y_test))
```

```python
# Random forest classifier

rfc=RandomForestClassifier()
rfc.fit(X_train,y_train)
print(rfc.score(X_test,y_test))
```
